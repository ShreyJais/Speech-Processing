# This lab experiment aims to apply the Fourier Transform on various signals in the time domain and analyze its frequency spectrum using the Discrete-Time Fourier Transform (DTFT) and the Discrete Fourier Transform (DFT).

#(1) Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

##(a) Generate a basic sinusoidal signal in the time domain.
(For example, you may generate a sine wave with a frequency of 5Hz,
sampled at 1000Hz.)

```python
n=5000
t_end = 5
t = np.linspace(0, t_end, n, endpoint=False)
frequency = 5
amplitude = 2
phase = 0

signal = amplitude * np.sin(2 * np.pi * frequency * t + phase)
sampling_rate = n/t_end
print(f'Sampling Rate :\t{sampling_rate}Hz')
```

## (b) Plot the time-domain waveform of the signal.

```python
px.line(x=t, y=signal, title='Sinusoidal Signal', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

##(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con-
tinuous frequency spectrum.

Discrete-Time Fourier Transform (DTFT) is used to make discrete time signal into continuous time signal

```python
def dtft(signal, sampling_rate):
  n = len(signal)
  f = np.fft.fftfreq(n, 1/sampling_rate)
  dtft_n = np.fft.fft(signal)/n
  px.line(x=f, y=np.abs(dtft_n), title='DTFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'}).show()
  return f, dtft_n
f, d = dtft(signal, sampling_rate)
#px.line(x=f, y=np.abs(d), title='DTFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'})
```

## (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
def dft(signal, sampling_rate):
  n = len(signal)
  f = np.fft.fftfreq(n, 1/sampling_rate)
  dft_n = np.fft.fft(signal)/n
  px.line(x=f, y=np.abs(dft_n), title='DFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'}).show()
  return f, dft_n

f, d = dft(signal, sampling_rate)
#px.line(x=f, y=np.abs(d), title='DFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'})
```

Both the DTFT and DFT graphs show a prominent peak at the frequency of +/-5 Hz which is the frequency content of the signal.

## (a) Generate a composite signal by adding two or more sinusoidal signals of different frequencies and amplitudes.

```python
n=5000
t_end = 5
t = np.linspace(0, t_end, n, endpoint=False)
f1 = 5
f2 = 10
a1 = 2
a2 = 4
phase1 = 0
phase2 = np.pi/2

signal1 = a1 * np.sin(2 * np.pi * f1 * t + phase1)
signal2 = a2 * np.sin(2 * np.pi * f2 * t + phase2)

composite_signal = signal1 + signal2
sampling_rate = n/t_end
print(f'Sampling Rate :\t{sampling_rate}Hz')
```

## (b) Plot the time-domain waveform of the composite signal.

```python
px.line(x=t, y=signal1, title='Sinusoidal Signal1', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

```python
px.line(x=t, y=signal2, title='Sinusoidal Signal2', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

```python
px.line(x=t, y=composite_signal, title='Composite Sinusoidal Signal', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

## (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.

```python
f, d = dtft(composite_signal, sampling_rate)
```

## (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.

```python
f, d = dft(composite_signal, sampling_rate)
```

> We know,
* frequency are at +/-(5,10)Hz
* Amplitud are at 2, 4

> from the DTFT and DFT graphs :
* Two Peaks: The presence of two peaks confirms the existence of two frequency components in the composite signal.
* Peak Locations: The locations of the peaks correspond to the frequencies of the constituent sine waves.
* Peak Amplitudes: The relative amplitudes of the peaks reflect the relative amplitudes of the sine waves.

## (a) Generate a rectangular signal.

```python
n = 5000
t_end = 5
t = np.linspace(0, t_end, n, endpoint=False)
pulse_width = 0.5
rectangular_signal = np.where((t >= 1) & (t <= 1 + pulse_width), 1, 0)
sampling_rate = n / t_end
print(f'Sampling Rate :\t{sampling_rate}Hz')
```

## (b) Plot the time-domain waveform.

```python
px.line(x=t, y=rectangular_signal, title='Rectangular Signal', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

## (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum.


```python
f, d = dtft(rectangular_signal, sampling_rate)
```

## (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.


```python
f, d = dft(rectangular_signal, sampling_rate)
```

## (e) Analyze the relationship between the time-domain waveform and the frequency-domain representation.

* Since the signal is not periodic, we don't expect to see a single, sharp peak.
* Since the signal has no frequency i.e 0 hence the position of pick is at 0hz
* The shape of the peak will depend on the pulse width. **A smaller pulse width will result in a wider peak, while a larger pulse width will result in a narrower peak.**
