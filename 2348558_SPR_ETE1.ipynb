```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

1

```python
n=5000
t_end = 1
t = np.linspace(0, t_end, n, endpoint=False)
frequency = 5
amplitude = 1
phase = 0

sinusoidal_signal = amplitude * np.sin(2 * np.pi * frequency * t + phase)
sampling_rate = n/t_end
print(f'Sampling Rate :\t{sampling_rate}Hz')
```

a

```python
px.line(x=t, y=sinusoidal_signal, title='Sinusoidal Signal', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

```python
sampling_rates = [20, 40, 100]
for sr in sampling_rates:
  indices = np.arange(0, len(t), int(sampling_rate / sr))
  discrete_signal = sinusoidal_signal[indices]
  discrete_t = t[indices]
  px.scatter(x=discrete_t, y=discrete_signal, title=f'Discrete Signal (Sampling Rate: {sr} Hz)', labels={'x': 'Time (s)', 'y': 'Amplitude'}).show()
  px.line(x=discrete_t, y=discrete_signal, title=f'Discrete Signal (Sampling Rate: {sr} Hz)', labels={'x': 'Time (s)', 'y': 'Amplitude'}).show()
```

c

Inference on Discrete Signal on sampling rate [20,40,100]Hz

* as the sampling rates reducing the audio data i.e the signal appears to be ploted by fewer data points and lose the continues curve and appears pointi and straight lines

* as the sampling rates increases the true shape of the audio data i.e the representation becomes more detailed and more samples are taken within a given time period allowing for a closer approximation of the continuous sinusoidal wave.

```python
f0 = 50  # Frequency of the sine wave (Hz)
fs = 1000  # Sampling rate (Hz)
N = 200  # Number of samples

n = np.arange(N)
sinusoidal_signal2 = np.sin(2 * np.pi * (f0 / fs) * n)
px.line(x=n, y=sinusoidal_signal2, title='Sinusoidal Signal of 200 samples', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

```python
def dtft(signal, sampling_rate):
  n = len(signal)
  f = np.fft.fftfreq(n, 1/sampling_rate)
  dtft_n = np.fft.fft(signal)/n
  px.line(x=f, y=np.abs(dtft_n), title='DTFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'}).show()
  return f, dtft_n
f, d = dtft(sinusoidal_signal2, fs)
#px.line(x=f, y=np.abs(d), title='DTFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'})
```

c

```python
def dft(signal, sampling_rate):
  n = len(signal)
  f = np.fft.fftfreq(n, 1/sampling_rate)
  dft_n = np.fft.fft(signal)/n
  px.line(x=f, y=np.abs(dft_n), title='DFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'}).show()
  return f, dft_n

f, d = dft(sinusoidal_signal2, fs)
#px.line(x=f, y=np.abs(d), title='DFT', labels={'x': 'Frequency (Hz)', 'y': 'Amplitude'})
```

c

Inference on the DTFT and DFT plots
* Similarities:
  * Both DTFT and DFT plots represent the frequency content of a signal as 50Hz.
  * Both have a similar shape and both have 2 picks in the frequency domain.
* Differences:
  * DTFT is a theoretical continuous-time Fourier transform of a discrete-time signal amd DFT is a discrete approximation of the DTFT.

```python
import numpy as np

x = np.array([1, 0, 0, 1, 0, 0, 1]) #(glottal pulse)
px.scatter(x=np.arange(len(x)),y=x, title='Glottal Pulse', symbol=x ,labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

b

```python
h = np.array([1, -0.5, 0.25, -0.125, 0.0625, -0.03125])
n = np.arange(len(h))
px.scatter(x=n,y=h, title='Vocal Track Model' ,labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

```python
y = np.convolve(x, h, mode='full')
px.line(x=np.arange(len(y)), y=y, title='Convolved Signal (Speech Signal)', labels={'x': 'Time (s)', 'y': 'Amplitude'})
```

The convolution between the source signal and the vocal tract impulse response simulates how the vocal tract shapes the raw glottal pulses into intelligible speech sounds. Specifically:

 - The source signal, which consists of simple impulses, is shaped by the impulse response of the vocal tract. This introduces resonances or formant frequencies that are characteristic of speech sounds.
 - The convolution result has a more complex structure compared to the original impulses, which now reflect the resonances introduced by the vocal tract model.

In summary, the filtering process (convolution) models how the vocal tract modifies the raw sound produced by the glottis to create speech sounds, highlighting the key role of the vocal tract in speech synthesis.
